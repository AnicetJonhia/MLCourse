```python
import matplotlib.pyplot as plt

plt.close("all")

import numpy as np
import pandas as pd
```

## Basic : plot

```python
np.random.seed(123456789)


ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2025", periods=1000))

ts = ts.cumsum()
ts.plot()
```

```python
df = pd.DataFrame(
    np.random.randn(1000, 4),
    index=ts.index,
    columns=list("ABCD")
)

df = df.cumsum()

plt.figure()
df.plot()
```

```python
# let's plot one column using x and y keywords

df3 = pd.DataFrame(
    np.random.randn(1000, 2),
    columns=["B", "C"]
)

df3["A"] = pd.Series(list(range(len(df))))

df3.plot(x="A", y="B")
```

```python
# plot.bar()
# using rand for getting positive numbers


df2 = pd.DataFrame(
    np.random.rand(10, 4),
    columns = list("abcd")
)
print(df2)
df2.plot.bar()


# stack bar plot
df2.plot.bar(stacked=True)
```

```python
# horizontal bar plot
# using barh

df2.plot.barh(stacked=True)
```

## Histograms

```python
df4 = pd.DataFrame(
     {
         "a": np.random.randn(1000) + 1,
         "b": np.random.randn(1000),
         "c": np.random.randn(1000) - 1
     },
     columns=list("abc")
)

df4
```

```python
# DataFrame.plot.hist()
plt.figure()
df4.plot.hist(alpha=0.5)
```

## Box plot

```python
df = pd.DataFrame(
    np.random.rand(10,5),
    columns=list("ABCDE")
)
df.plot.box()
```

## Are plot

```python
df = pd.DataFrame(np.random.rand(10, 4), columns=["a", "b", "c", "d"])
df.plot.area()
```

## Scatter plot

```python
df = pd.DataFrame(np.random.rand(50, 4), columns=["a", "b", "c", "d"])

df["species"] = pd.Categorical(
    ["setosa"] * 20 + ["versicolor"] * 20 + ["virginica"] * 10

  )

df.plot.scatter(x="a", y="b")


df.plot.scatter(x="a", y="b", c="species", cmap="viridis", s=50);
```

## Hexagon bin plot

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=["a", "b"])

df["b"] = df["b"] + np.arange(1000)

df.plot.hexbin(x="a", y="b",gridsize=25)
```

## Pie plot

```python
series = pd.Series(
    3 * np.random.rand(4),
    index=["a", "b", "c", "d"],
    name="series"
)

series.plot.pie(figsize=(6, 6))
```

```python
# with a data frame
df = pd.DataFrame(
    3 * np.random.rand(4,3), index=["a", "b", "c", "d"], columns=["x","y","z"]
)
df.head()

df.plot.pie(subplots=True,figsize=(12, 8), autopct="%.2f")
```

## Scatter matrix plot

```python
from pandas.plotting import scatter_matrix

df = pd.DataFrame(np.random.rand(1000, 4), columns=list("abcd"))
scatter_matrix(df, alpha=0.2, figsize=(6, 6), diagonal="kde")
```
